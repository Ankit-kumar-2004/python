```python
def factorial(n):
    if n == 1:      # Base case
        return 1
    else:
        return n * factorial(n-1)   # Recursive call

print(factorial(5))
```

```python
def show(n):
    if n==0:
        print("base case reached")
        return   
    else:
        print(n)
        show(n-1)
show(5)
```

```python
def sum_natural_numbers(n):
    if n==1:
        return 1
    else:
        return n + sum_natural_numbers(n-1)
print(sum_natural_numbers(5))
```

```python
def print_list(lst,idx):
    if idx==len(lst):
        return
    else:
        print(lst[idx])
        print_list(lst,idx+1)
fruits=["apple","banana","cherry"]
print_list(fruits,idx=0)
```

```python
def sum_of_list(lst,idx):
    if idx==len(lst):
        return 0
    else:
        return lst[idx] + sum_of_list(lst,idx+1)
numbers=[1,2,3,4,5]
print(sum_of_list(numbers,idx=0))
```
